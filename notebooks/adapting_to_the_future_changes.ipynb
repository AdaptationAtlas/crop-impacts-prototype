```python
import pandas as pd
import numpy as np
import scipy.stats
```

```python
df = pd.read_csv("granular_yields_fixed_new_maize_data.csv")
df.head()
```

```python
def median_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.median(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h
```

```python
adapting_to_future_changes_options = [
    {"cultivar": "current", "yield": "yield_rfd", "label": "No adaptation"},
    {"cultivar": "current", "yield": "yield_irr", "label": "Current cultivar, irrigation"},
    {"cultivar": "early", "yield": "yield_rfd", "label": "Early maturity, no irrigation"},
    {"cultivar": "early", "yield": "yield_irr", "label": "Early maturity, irrigation"},
    {"cultivar": "late", "yield": "yield_rfd", "label": "Late maturity, no irrigation"},
    {"cultivar": "late", "yield": "yield_irr", "label": "Late maturity, irrigation"}
]
```

```python
# select current cultivar, irrigation
#option = adapting_to_future_changes_options[2]
crops = df["crop"].unique()
admins = ["admin0_name", "admin1_name", "admin2_name"]

cols_to_group_by = ["cycle", "ssp"]
horizons = [2005, 2030, 2050]

columns = [
    "cycle",
    "ssp",
    "adaptation",
    "cultivar",
    "yield",
    "crop",
    "horizon",
    "mean_difference",
    "ci_upper",
    "ci_lower",
    "admin0_name",
    "admin1_name",
    "admin2_name",
]

rows = []

for i in range(1, len(admins)+1):
    levels = admins[:i]
    print(levels)
    none_levels = admins[i:]

    for group, groupdf in df.groupby(cols_to_group_by + levels):
        for c in crops:
            for option in adapting_to_future_changes_options:
                filtered = groupdf[
                    (groupdf["cultivar"] == option["cultivar"]) &
                    (groupdf["crop"] == c)
                ]
                value_col = f"wmean_{option['yield']}"
                values = filtered#[[value_col]]
                for h in horizons:
                    median, cil, cih = median_confidence_interval(
                        values[values["horizon"] == h][value_col]
                    )
                    if h == 2005:
                        h = "historical"
                    rows.append([
                        group[0], group[1],
                        option["label"], option["cultivar"], option["yield"], c, h, median, cil, cih,
                    ] + list(group[2:]) + [None] * len(none_levels))
    #        break
    #    break
    #break

values
```

```python
adaptation_data = pd.DataFrame(rows, columns=columns)
adaptation_data
```
